```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
def nothing(x):
    pass
```

```python
# 마우스 동작
drawing = False
mode = True
ix,iy = 0, 0
cx, cy = -1, -1
org_img = cv2.imread('mt.jpg')
img = org_img.copy()

def draw_circle(event, x, y, flags, param):
    global ix, iy, cx, cy, drawing, org_img
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        cx, cy = x, y
        
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        # 드래그 방향에 상관없이 영역을 잡기 위해 min, max 사용
        x_start, x_end = min(ix, x), max(ix, x)
        y_start, y_end = min(iy, y), max(iy, y)
        
        # 반투명 붉은색 효과: 해당 영역의 B(0), G(1) 채널을 0으로 설정
        # (이미지 데이터가 있을 때만 실행되도록 범위 체크)
        if x_end > x_start and y_end > y_start:
            org_img[y_start:y_end, x_start:x_end, 0:2] = 0
```

```python
def nothing(x):
    pass

cv2.namedWindow('image')
cv2.createTrackbar('value', 'image', 0, 255, nothing)
cv2.setMouseCallback('image', draw_circle)

font = cv2.FONT_HERSHEY_SIMPLEX

while(1):
    img = org_img.copy()
    value = cv2.getTrackbarPos('value','image')
    
    # 드래그 중일 때 실시간으로 사각형 가이드라인 표시 (선택 사항)
    if drawing:
        cv2.rectangle(img, (ix, iy), (cx, cy), (0, 255, 0), 1)
    
    cv2.putText(img, f"Mouse Position : ({ix}, {iy}) - ({cx}, {cy}) - {value}", 
                (10,30), font, 0.5, (255,255,255), 1, cv2.LINE_AA)
    
    cv2.imshow('image', img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27: # ESC
        break

cv2.destroyAllWindows()

cv2.destroyAllWindows()
```
